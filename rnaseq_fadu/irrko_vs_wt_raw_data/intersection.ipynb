```python
# Process and genes of the neo-loops

collagen_trimerization = ["COL11A1", "COL16A1", "COL1A1", "COL12A1", "COL9A1", "COL21A1", "COL5A1", "COL4A2", "COL2A1", "COL6A1", "COL6A2", "COL8A1", 
                          "COL26A1", "COL18A1", "COL4A1", "COL14A1"]

cornified = ["PRSS8", "DSP", "CSTA", "SPINK5", "PCSK6", "PKP4", "LCE2B", "SPRR2G", "IVL", "SPRR3", "SPRR2D", "KRT1", "KRT24", "SPRR1B", "SPRR1A", 
             "KRT78", "KRT4", "KRT72", "KRT27", "LCE1D", "KRT2", "KRT76", "KRT79", "KRT73", "LCE1E", "KRT26", "KRT3", "LCE1A", "LCE4A", "LCE2A", 
             "LCE2C", "LCE2D", "KRT77", "LCE1B", "SPRR2B", "LCE1C", "LORICRIN", "LELP1", "SPRR2E", "KRT25", "LCE6A", "LCE1F", "SPRR2A", "SPRR2F"]

collagen_enzymes = ["TLL1", "COL11A1", "COL16A1", "P3H2", "COL1A1", "COL12A1", "COL9A1", "COL21A1", "COL5A1", "COL4A2", 
                    "ADAMTS14", "COL2A1", "COL6A1", "COL6A2", "COL8A1", "COL26A1", "PCOLCE2", "COL18A1", "COL4A1", "COL14A1"]

keratinization = ["PRSS8", "DSP", "CSTA", "SPINK5", "PCSK6", "PKP4", "LCE2B", "SPRR2G", "IVL", "SPRR3", "SPRR2D", "KRT1", "KRT24", "SPRR1B", "SPRR1A", "KRT78", "KRT4", "KRT72", "KRT27", "LCE1D", "KRT2", 
                  "KRT76", "KRT79", "KRT73", "LCE1E", "KRT26", "KRT3", "LCE1A", "LCE4A", "LCE2A", "KRTAP12-1", "LCE2C", "LCE2D", "KRTAP10-8", "KRTAP10-6", "KRTAP10-12", "KRT77", "LCE1B", "SPRR2B", 
                  "LCE1C", "LORICRIN", "LELP1", "SPRR2E", "KRT25", "KRTAP12-3", "KRTAP10-2", "KRTAP12-4", "KRTAP10-3", "KRTAP10-4", "KRTAP10-1", "KRTAP10-9", "KRTAP10-10", "KRTAP12-2", 
                  "LCE6A", "LCE1F", "KRTAP10-5", "SPRR2A", "KRTAP10-11", "SPRR2F", "KRTAP10-7"]

```

```python
# Number of genes per list
print("Number of genes per list:")
print("Collagen trimerization:", len(collagen_trimerization))
print("Cornified envelope:", len(cornified))        
print("Collagen catabolic enzymes:", len(collagen_enzymes))
print("Keratinization:", len(keratinization))
```

```python
# Join all gene list into one and remove duplicates
genes_neoloops = list(set(collagen_trimerization + cornified + collagen_enzymes + keratinization))
set1 = set(genes_neoloops)
print(f"Number of genes in Neo-loops: {len(genes_neoloops)}")
```

```python
# Genes in FaDu RNAseq cell line WT vs IrrKO

genes_rnaseq_wt = ["ZC3HAV1", "DTX3L", "ZNFX1", "IFITM1", "SLC7A11", "TRIM14", "PARP14", 
                   "UBE2L6", "TNFSF10", "TRIM22", "BTN3A1", "DDX60L", "SAMD9L", "SAMD9", 
                   "APOL6", "APOL3", "IFIH1", "WARS", "OAS3", "ACKR3", "HCAR2", "GRB10", 
                   "PMAIP1", "IRF9", "APOL1", "HLA-B_3", "PLCG2", "DHX58", "PODXL", "VEGFC", 
                   "KLF4", "TMEM173", "ETV7", "PPM1K", "IFI6", "PRR15", "SP8", "HERC6", "EPSTI1", 
                   "BST2", "ISG20", "ISG15", "NCF2", "OAS1", "SPOCK1", "RARRES3", "FGF2", "CYP2W1", 
                   "RTP4", "RNF144A", "IFIT1", "DDX58", "MX1", "MX2", "XAF1", "PTPRZ1", "CMPK2", 
                   "GBP4", "OAS2", "IFI44", "IFI44L", "IFIT3", "DDX60", "IFIT2", "HERC5", "SCN3A",
                    "OASL", "RSAD2"
    
]

# Size of the transcriptome
# Tama√±o del transcriptoma IrrWT vs Control FaDu
print(len(genes_rnaseq_wt))  # 68 genes 
```

```python
genes_rnaseq_irrko = ["MROH2A", "FNDC7", "IL1R2", "MMP3", "SPRR1A", "LOC101928841", "TCHHL1", 
    "MMP10", "IGFL1", "AMPD1", "IL1RL1", "ANPEP", "MMP1", "STC1", "MMP12", ""
    "SPRY4", "MMP13", "CDH11", "PADI1", "KRT81", "SPRR1B", "CNGB1", "SLC12A7_1", 
    "ODC1", "RPLP0P2", "FAM65C", "CD24", "CST6", "KIAA1199", "MAP1B", "FST", "GPRC5B", 
    "RAB3B", "PXDN", "LIF"
]

# Size of the transcriptome IrrKO vs Control FaDu
print(len(genes_rnaseq_irrko))  # 35
```

```python
# Antibody list
genes_ab = [
    "ITGB3", "MMP1", "MMP2", "MMP7", "MMP8", "MMP11", "MMP13", "MMP14",
    "MMP17", "MMP19", "MMP28", "TIMP3", "COL1A1", "FN1"
]

print(len(genes_ab))  # 14 genes
```

```python
# Make list to contains 
set1 = set(genes_neoloops)
set2 = set(genes_rnaseq_irrko)
set3 = set(genes_ab)

# PLot Venn diagram
plt.figure(figsize=(8,8))
venn3([set1, set2, set3], set_labels=('Genes Neo-loops', 'Genes RNA-seq', 'Antibodies'))

# Show plot
plt.title("Intersection of genes between the two methodologies and antibodies")
plt.show()
```

```python
# Imprimir las intersecciones
print("Intersection between Neo-loops and RNA-seq:", set1.intersection(set2))
print("Intersection between Neo-loops and Antibodies:", set1.intersection(set3))
print("Intersection between RNA-seq and Antibodies:", set2.intersection(set3))
print("Intersection between the three sets:", set1.intersection(set2).intersection(set3))
```
